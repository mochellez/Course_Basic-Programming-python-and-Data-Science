```python
#Menangani dalam kondisi tertentu

try:
  angka=int(input('Masukkan angka:'))
  hasil=100/angka
except ZeroDivisionError:
  print('Tidak boleh pembagian nol')

except ValueError:
  print('Tidak valid')
```

```python
#Menggunakan else
try:
  angka=int(input('Masukkan angka:'))
  hasil=100/angka
except ZeroDivisionError:
  print('Tidak boleh pembagian nol')

except ValueError:
  print('Tidak valid')

else:
  print('Hasil=',hasil)
```

```python
#Menggunakan finally
try:
  angka=int(input('Masukkan angka:'))
  hasil=100/angka
except ZeroDivisionError:
  print('Tidak boleh pembagian nol')

except ValueError:
  print('Tidak valid')

else:
  print('Hasil=',hasil)

finally:
  print('Program selesai')
```

```python
#Modul Kalkulator

%%writefile kalkulator.py
def tambah(a,b):
  return a + b
def kurang(a,b):
  return a - b
def kali(a,b):
  return a * b
def bagi(a,b):
  return a / b
```

```python
import kalkulator

print('Penjumlahan :', kalkulator.tambah(10,20))
print('Pengurangan :', kalkulator.kurang(60,40))
print('Perkalian :', kalkulator.kali(60,40))
print('Pembagian :', kalkulator.bagi(60,40))
```

```python
%%writefile bangundatar.py
pi = 3,14
def luas_persegi(sisi):
  return sisi*sisi
def luas_persegi_panjang(panjang,lebar):
  return panjang*lebar
def luas_lingkaran(pi,jarijari):
  return pi*jarijari*jarijari
```

```python
import bangundatar

print('Persegi :', bangundatar.luas_persegi(4))
print('Persegi panjang :', bangundatar.luas_persegi_panjang(8,4))
print('Lingkaran :', bangundatar.luas_lingkaran(4,4))
```
