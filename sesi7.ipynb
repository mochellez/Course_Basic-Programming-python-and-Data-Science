```python
def luas_persegi(sisi):
  return sisi*sisi

print(luas_persegi(5))
print(luas_persegi(7))
print(luas_persegi(10))
```

```python
#Contoh fungsi tanpa parameter dan tanpa return
def salam():
  print('Selamat datang di program python!')
  print('Semoga belajarmu menyenangkan.')

salam()
```

```python
#Contoh fungsi dengan parameter dan return
def luas_segitiga(alas,tinggi):
  luas = (alas*tinggi) / 2
  return luas

#Memanggil fungsi
hasil = luas_segitiga(10,5)
print(f'Luas segitiga: {hasil}')

hasil2 = luas_segitiga(8,3)
print(f'Luas segitiga: {hasil2}')
```

```python
#Contoh fungsi dengan default parameter
def perkenalan(nama='NaN', asal='tidak diketahui', umur=0):
  print(f'Nama : {nama}')
  print(f'Asal : {asal}')
  print(f'Umur : {umur} tahun')
  print('---')

#Memanggil dengan semua argumen
perkenalan('Budi','Medan',20)

#Memanggil hanya nama saja, sisanya pakai default
perkenalan('Sari')

perkenalan('Michelle','',16)
```

```python
#Contoh fungsi dengan multiple return value
def operasi_matematika(a,b):
  jumlah = a+b
  selisih = a-b
  kali = a*b
  bagi = a/b
  return jumlah,selisih,kali,bagi

#Menangkap semua nilai return
tambah,kurang,kali,bagi = operasi_matematika(20,4)

print(f'Penjumlahan : {tambah}')
print(f'Pengurangan : {kurang}')
print(f'Perkalian : {kali}')
print(f'Pembagian : {bagi}')
```

```python
#Contoh fungsi memanggil fungsi lain
def celcius_ke_fahrenheit(c):
  return (c*9/5)+32

def celcius_ke_kelvin(c):
  return c+272.15

def konversi_suhu(celcius):
  f = celcius_ke_fahrenheit(celcius)
  k = celcius_ke_kelvin(celcius)
  print(f'Suhu : {celcius} C')
  print(f'Fahrenheit : {f} F')
  print(f'Kelvin : {k} K')

#Cukup panggil satu fungsi utama
konversi_suhu(100)
print('---')
konversi_suhu(0)
```

```python
#Contoh class Dasar
class Mahasiswa:
  #Constructor - dipanggil saat objek dibuat
  def __init__(self,nama,nim,jurusan):
    self.nama = nama
    self.nim = nim
    self.jurusan = jurusan
    self.nilai = [] # list kosong untuk menyimpan nilai

  #Method - menambahkan nilai
  def tambah_nilai(self,nilai):
    self.nilai.append(nilai)
    print(f'Nilai{nilai} berhasil ditambahkan.')

  #Method - menghitung rata-rata
  def rata_rata(self):
    if len(self.nilai)==0:
      return 0
      return sum(self.nilai) / len(self.nilai)
  def peringkat(self):
    rata_rata = self.rata_rata()
    if rata_rata >= 90:
        return 'A'
    elif rata_rata >= 80:
        return 'B'
    elif rata_rata >= 70:
        return 'C'
    else:
        return 'D'
  #Method - menampilkan info mahasiswa
  def tampilkan_info(self):
    print(f'========================')
    print(f'Nama : {self.nama}')
    print(f'Nim : {self.nim}')
    print(f'Jurusan : {self.jurusan}')
    print(f'Nilai : {self.nilai}')
    print(f'Rata-rata : {self.rata_rata():.2f}')
    print(f'Peringkat : {self.peringkat()}')
    print(f'========================')

  #Method object/instance
  mhs1 = Mahasiswa('Budi','2101001','Informatika')
  mhs2 = Mahasiswa('Sari','2101002','Sistem Informasi')
  mhs3 = Mahasiswa('Michelle','2101003','Informatika')

  #Menggunakan method
  mhs1.tambah_nilai(85)
  mhs1.tambah_nilai(90)
  mhs1.tambah_nilai(78)
  mhs1.tampilkan_info()

  mhs2.tambah_nilai(92)
  mhs2.tambah_nilai(88)
  mhs2.tampilkan_info()

  mhs3.tambah_nilai(90)
  mhs3.tambah_nilai(89)
  mhs3.tambah_nilai(86)
  mhs3.tampilkan_info()
```
