```python
#Struktur if else
hujan = False

if hujan:
  print('Bawa payung')
else:
  print('Tidak perlu bawa payung')


meteor = True

if meteor:
  print('Lompat!')
else:
  print('Lari terus!')


kena_koin = False

if kena_koin:
  print('+10 skor')
else:
  print('Skor tetap')


punya_id = True

if punya_id:
  print('Silahkan masuk')
else:
  print('Akses ditolak')


password = 'python123'

if password == 'python123':
  print('Login berhasil!')
else:
  print('Password salah')
```

```python
#Struktur if elif else
nilai = 82
if nilai >= 90:
  print('Grade A')
elif nilai >= 75:
  print('Grade B')
elif nilai >= 60:
  print('Grade C')
else:
  print('Perlu Remedial')
```

```python
#Operator and
nilai = 85
kehadiran = 96

if nilai >= 75 and kehadiran >= 80:
  print('Boleh mengikuti ujian')
```

```python
nilai = 85
kehadiran = 90

if nilai >= 75 and kehadiran >= 80:
  print('Selamat, Anda Lulus!')
else:
  print('Belum Lulus')
```

```python
username = 'admin'
password = 'python123'

if username == 'admin' and password == 'python123':
  print('Login Berhasil')
else:
  print('Username atau Password Salah')
```

```python
#Operator Or
nilai = 85
kehadiran = 96

if nilai >= 90 or kehadiran >= 95:
  print('Boleh mengikuti lomba')
else:
  print('Belum memenuhi syarat')


email = False
username = True

if email or username:
  print('Login berhasil')
else:
  print('Login gagal')
```
