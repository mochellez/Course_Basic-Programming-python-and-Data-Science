```python
import numpy as np
nilai = np.array([67,68,70,65,90,97,77,45,75])
kkm = 70

lulus = np.sum(nilai >= kkm)
print(lulus)
tidak_lulus = np.sum(nilai < kkm)
print(tidak_lulus)
nilai_terbesar_kedua = list(set(nilai))
nilai_terbesar_kedua.sort(reverse = True)
print(nilai_terbesar_kedua[3])
```
