```python
def login(username,password):
  return username == 'admin' and password == '1234'

def pilih_barang():
  return 'Laptop'

def checkout(barang):
  return f'Checkout berhasil untuk {barang}'

#System Test
if login('admin','1234'):
  barang = pilih_barang()
  hasil = checkout(barang)
  print(hasil)
else:
  print('Login Gagal')
```

```python
def login(username, pw):
  data_user ={'Michelle':'1234','admin':'admin123'}

#Acceptance Test
  if username not in data_user:
    return 'User tidak ditemukan'
  if data_user[username]!=pw:
    return 'Password Salah'

  return 'Login Berhasil'
```

```python
assert login('Michelle','1234') == 'Login Berhasil'
assert login('pengguna','1234') == 'User tidak ditemukan'
assert login('Michelle','12345') == 'Password Salah'
print('Semua Unit Test berhasil!')
```

```python
def cek_kelulusan(nilai):
  if nilai >= 75:
    return 'Lulus'
  return 'Tidak Lulus'

#Black Box Test
print(cek_kelulusan(90))
print(cek_kelulusan(60))
```

```python
def cek_kelulusan(nilai):
  if nilai >= 75:
    return 'Lulus'
  else:
    return 'Tidak Lulus'

#White Box Test
assert cek_kelulusan(80) == 'Lulus'  #Jalur True
assert cek_kelulusan(50) == 'Tidak Lulus'  #Jalur False
print('Semua jalur program telah diuji')
```

```python
# Versi lama
def login(username,password):
  return username == 'admin' and password == '1234'

assert login('admin','1234') == True

#Menambahkan fitur baru
def lupa_password(email):
  return 'Link reset password telah dikirim'

#Regression Test
assert login('admin','1234') == True
print(lupa_password('admin@email.com'))

print('Fitur lama tetap berjalan')
```
