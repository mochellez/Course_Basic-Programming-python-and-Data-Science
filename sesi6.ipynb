```python
#Line Chart Matplotlib (menampilkan tren atau pola)
import matplotlib.pyplot as plt

bulan = ['Jan','Feb','Mar','Apr','Mei']
penjualan = [100,120,150,130,180]

plt.plot(bulan,penjualan)
plt.title('Grafik Penjualan')
plt.xlabel('Bulan')
plt.ylabel('Jumlah Penjualan')
plt.show()
```

```python
#Bar Chart Matplotlib (melakukan perbandingan)
import matplotlib.pyplot as plt

produk = ['Laptop','Mouse','Keyboard','Monitor']
terjual = [25,40,30,20]

plt.bar(produk,terjual)

plt.title('Jumlah Produk Terjual')
plt.xlabel('Produk')
plt.ylabel('Jumlah')
plt.show()
```

```python
#Pie Chart (melihat persentase)
import matplotlib.pyplot as plt
label = ['Python','Java','C++','JavaScript']
jumlah = [45,25,15,15]

plt.pie(jumlah,labels=label,autopct='%1.1f%%')

plt.title('Bahasa Pemrograman Favorit')
plt.show()
```

```python
#Scatter Plot (melihat hubungan antara dua variabel)
import matplotlib.pyplot as plt

jam_belajar = [1,2,3,4,5,6]
nilai = [50,60,68,75,85,95]

plt.scatter(jam_belajar,nilai)

plt.title('Hubungan Jam Belajar dan Nilai')
plt.xlabel('Jam Belajar')
plt.ylabel('Nilai')

plt.show()
```

```python
#Histogram
import matplotlib.pyplot as plt

nilai = [70,75,80,80,82,85,87,90,90,91,92,95]

plt.hist(nilai)

plt.title('Distribusi Nilai Siswa')
plt.xlabel('Nilai')
plt.ylabel('Frekuensi')

plt.show()
```

```python
#Map
import matplotlib.pyplot as plt
import cartopy.crs as ccrs

fig = plt.figure(figsize=(8,6))
ax = plt.axes(projection=ccrs.PlateCarree())

ax.coastlines()
ax.set_title('Peta Dunia')
plt.show()
```

```python
#Bar Chart seaborn
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({'Produk':['Laptop','Mouse','Keyboard','Monitor'],'Terjual':[50,120,80,40]})
sns.barplot(data=df,x='Produk',y='Terjual')
plt.title('Jumlah Produk Terjual')
plt.show()
```

```python
#Line Chart
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({'Bulan':['Jan','Feb','Mar','Apr','Mei'],'Penjualan':[100,120,150,130,180]})
sns.lineplot(data=df,x='Bulan',y='Penjualan',marker='o')
plt.title('Grafik Penjualan')
plt.show()
```

```python
#Scatter Plot
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({'Jam Belajar':[1,2,3,4,5,6],'Nilai':[50,60,70,80,90,95]})
sns.scatterplot(data=df,x='Jam Belajar',y='Nilai',s=100)
plt.title('Jam Belajar vs Nilai')
plt.show()
```

```python
#Histogram
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

nilai = [70,72,74,75,78,80,82,84,85,86,88,90,92,95]

sns.histplot(nilai,bins=6,kde=True)

plt.title('Distribusi Nilai')
plt.show()
```

```python
#Box Plot
import seaborn as sns
import matplotlib.pyplot as plt

# Load datase bawaan Seaborn
tips = sns.load_dataset('tips')

# Membuat Box Plot
sns.boxplot(y=tips['total_bill'])

plt.title('Distribusi Total Tagihan Pelanggan')
plt.ylabel('Total BIll ($)')

plt.show()
```
