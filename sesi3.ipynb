LIST

```python
#Mengakses Elemen List
buah = ['Apel','Jeruk','Mangga']
print(buah[0])
```

```python
#Menambah Data pada List
buah.append('Pisang')
print(buah)

#Menghapus Data pada List
buah.remove('Pisang')
print(buah)
```

```python
#Mengubah Data pada List
buah[0] = 'Semangka'
print(buah)

#Mengecek Panjang Data pada List
len(buah)
print(buah)
```

DICTIONARY

```python
#Mengakses Data Dictionary
mahasiswa = {'nama':'Andi','umur':20,'prodi':'Informatika'}
mahasiswa['nama'] = 'Ben'

#Menambahkan Data Dictionary
mahasiswa['ipk']= 3.85
mahasiswa['jenis kelamin']= 'laki-laki'

#Mengubah Data Dictionary
mahasiswa['umur'] = 21
mahasiswa['prodi'] = 'Manajemen'

#Menghapus Data Dictionary
del mahasiswa['umur']
print(mahasiswa)
```
