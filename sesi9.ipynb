INHERITANCE

```python
class Hewan: #Parent Class
  def makan(self):
    print('Hewan sedang makan')

class Kucing(Hewan): #Child Class
  def suara(self):
    print('Meong')
class Anjing(Hewan): #Child Class
  def suara(self):
    print('Guk')
class Burung(Hewan): #Child Class
  def suara(self):
    print('Kicau')

k = Kucing()
k.makan()
k.suara()

a = Anjing()
a.makan()
a.suara()

b = Burung()
b.makan()
b.suara()
```

POLYMORPHINS

```python
class Kendaraan:
  def bergerak(self):
    pass

class Mobil(Kendaraan):
  def bergerak(self):
    print('Mobil bergerak di jalan raya')

class Pesawat(Kendaraan):
  def bergerak(self):
    print('Pesawat terbang di udara')

class Kapal(Kendaraan):
  def bergerak(self):
    print('Kapal bergerak di laut')

kendaraan = [
    Mobil(),
    Pesawat(),
    Kapal()
]

for k in kendaraan:
  k.bergerak()
```

```python
class Pembayaran:
  def bayar(self,jumlah):
    pass

class TransferBank(Pembayaran):
  def bayar(self,jumlah):
    print(f'Transfer bank sebesar Rp{jumlah:,}')

class QRIS(Pembayaran):
  def bayar(self,jumlah):
    print(f'Scan QRIS sebesar Rp{jumlah:,}')

class KartuKredit(Pembayaran):
  def bayar(self,jumlah):
    print(f'Pembayaran kartu kredit sebesar Rp{jumlah:,}')

metode = [
    TransferBank(),
    QRIS(),
    KartuKredit()
]

for m in metode:
  m.bayar(50000)
```

```python
class Character:
  def attack(self):
    pass

class Warrior(Character):
  def attack(self):
    print('Warrior menyerang dengan pedang')

class Archer(Character):
  def attack(self):
    print('Archer menyerang dengan panah')

class Mage(Character):
  def attack(self):
    print('Mage menyerang dengan bola api')


tim = [
    Warrior(),
    Archer(),
    Mage()
]

for hero in tim:
  hero.attack()
```

```python
def hitung_mundur(n):
  if n == 0:
    print('Selesai')
    return

  print(n)
  hitung_mundur(n-1)

hitung_mundur(20)
```

```python
def faktorial(n):
  if n == 1:
    return 1

  return n * faktorial(n-1)

hasil = faktorial(4)
print(hasil)
```

```python
def jumlah(n):
  if n == 1:
    return 1

  return n + jumlah(n-1)

print(jumlah(5))
```

```python
def balik(teks):
  if len(teks) == 0:
    return ''

  return balik(teks[1:]) + teks[0]

print(balik('michelle'))
```
